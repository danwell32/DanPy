```python
accuracyvector= []
for n in range(1, 289):
    for m in range()
    currenttime = 'Time' + str(n)
    X = df[[currenttime, 'Electrode']]
    y = df[['Label']]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0, stratify=y)
    clf = SVC(C= 100, gamma= 0.01, kernel='rbf')
    clf.fit(X_train, y_train.values.flatten())
    predictions = clf.predict(X_test)
    print(accuracy_score(y_test, predictions))
    accuracyvector.append(accuracy_score(y_test, predictions))
print('media:',np.array(accuracyvector).mean(),'std:',np.array(accuracyvector).std())
```

```python
accuracyvector= []
for n in range(1, 289):
    currenttime = 'Time' + str(n)
    for m in range(0,30):
        print(currenttime + '  ' + 'Electrodo'+str(m))
        X = df.query('{} == Electrode'.format(m))
        X_ = X[[currenttime]]
        X_ = X_.to_numpy()
        y = df.query('{} == Electrode'.format(m))
        y = y[['Label']]
        y = y.to_numpy()
        y = y + 2*m
        #print(X_,y)
        X_train, X_test, y_train, y_test = train_test_split(X_, y.ravel(), test_size=0.25, random_state=0, stratify=y)
        clf = SVC(C= 1000, gamma= 0.1, kernel='rbf')
        clf.fit(X_train, y_train)
        predictions = clf.predict(X_test)
        accuracyvector.append(accuracy_score(y_test, predictions))
print('media:',np.array(accuracyvector).mean(),'std:',np.array(accuracyvector).std())
```
